```python
import json
import pandas as pd
from google.colab import files
```

```python
# 1- Load the file
uploaded = files.upload()

# 2- Read the file line-by-line and store each JSON object in a list

test_set = list(uploaded.keys())[0]

data = []
with open(test_set, 'r', encoding='utf-8') as file:
    for line in file:
        data.append(json.loads(line))

# 3- Convert to a DataFrame
df = pd.DataFrame(data)
```

```python
#Inspect the Data
print(df.head())
print("-" * 50)
print(df.info()) # overall structure
print("-" * 50)
print(df.shape) # Shape (Rows, Columns)
print("-" * 50)
print(df.isnull().sum()) # check for missing values
print("-" * 50)
print(df.describe()) # summary statistics
```

```python
#Inspect a random  Sample of Rows
random_samples = df.sample(n=25)

for i, row in random_samples.iterrows():
    print(f"Sample {i+1}:")
    print(f"Question: {row['question']}")
    print(f"Answer: {row['answer']}\n")
    print("-" * 50)

```
